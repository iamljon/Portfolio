```python
# Imports

import numpy  as np
import os
import pandas as pd
import quandl
import ta
```

```python
# load api key and features from config file

curr_working_dir = os.getcwd()
parent_path = os.path.abspath(os.pardir)
quandl_api_key = os.path.join(parent_path, 'cfg', 'quandl_api_key')

with open(quandl_api_key, "r") as file:
    api_key = file.readlines()[0]

metadata_file = 'FRED_metadata.csv'   
metadata_path = os.path.join(parent_path, 'cfg', metadata_file)
features = pd.read_csv(metadata_path, usecols=['code'])

feature_count = features.count()[0]
```

```python
# get data 

data = pd.DataFrame()

quandl.ApiConfig.api_key = api_key

for feature in range(feature_count):
    data[[features['code'][feature]]] = quandl.get( \
                                                    'FRED/{}'.format(features['code'][feature]), \
                                                    start_date='2000-01-01', end_date='2018-09-30')
```

```python
data_without_na = data.dropna(axis = 1)

pct_change = data_without_na.pct_change()

print(data_without_na)
print(pct_change)
```

```python
df = pd.DataFrame(pct_change, index = pd.date_range(start = '2000-01-01', end = '2018-08-31'))

print(df)
df.ffill(inplace=True)
df.bfill(inplace=True)
print(df)
```
